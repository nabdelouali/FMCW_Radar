```python
import numpy as np
import matplotlib.pyplot as plt

# Define the color palette
colors = {
    'black': '#000000',
    'yellow': '#F3E079',
    'soft_grey': '#D9D9D9',
    'muted_blue': '#A7BBC7',
    'light_green': '#C7E5A6',
    'white': '#FFFFFF',
    'beige': '#F5F2E0'
}

# Time parameters
t_start = 0      # Start time
t_end = 40       # End time
duration = t_end - t_start  # Duration
T_chirp = duration           # Chirp duration

# Chirp parameters
f_start = 0    # Start frequency in GHz
f_end = 20      # End frequency in GHz
B = f_end - f_start  # Bandwidth in GHz
S = B / T_chirp  # Chirp slope in GHz/µs

# Sampling frequency and number of samples
fs = 1000  # Set a fixed sampling rate for better visualization
num_samples = int(fs * T_chirp)

# Time vector
t = np.linspace(t_start, t_end, num_samples)  # Time in µs

# Generate the chirp signal
signal = np.cos(2 * np.pi * (f_start * t + 0.5 * S * t**2))

# Instantaneous frequency calculation
instantaneous_frequency = f_start + S * t

# Create the figure with a beige background
plt.figure(figsize=(18, 10), facecolor=colors['beige'])

# Plot 1: Amplitude vs Time
ax1 = plt.subplot(3, 1, 1, facecolor=colors['soft_grey'])
ax1.plot(t, signal, color=colors['muted_blue'])  # Time in µs
ax1.set_xlabel('Time (µs)', color=colors['black'])
ax1.set_ylabel('Amplitude', color=colors['black'])
ax1.set_title('Amplitude vs Time', color=colors['black'])
ax1.set_xlim(t_start, t_end)
ax1.set_ylim(-1, 1)
ax1.tick_params(axis='x', colors=colors['black'])
ax1.tick_params(axis='y', colors='black')
ax1.grid(True, color=colors['white'])

# Define the first zoom region (10 to 10.5 µs)
zoom_start1 = 10
zoom_end1 = 10.5
zoom_indices1 = np.where((t >= zoom_start1) & (t <= zoom_end1))
t_zoom1 = t[zoom_indices1]
signal_zoom1 = signal[zoom_indices1]
ax1.axvspan(zoom_start1, zoom_end1, color=colors['light_green'], alpha=0.3)

# Plot 2: Amplitude vs Time (Zoomed-In for 10 to 10.5 µs)
ax2 = plt.subplot(3, 3, 4, facecolor=colors['soft_grey'])
ax2.plot(t_zoom1, signal_zoom1, color=colors['muted_blue'])
ax2.set_xlabel('Time (µs)', color=colors['black'])
ax2.set_ylabel('Amplitude', color=colors['black'])
ax2.set_title('Zoomed-In (10 to 10.5 µs)', color=colors['black'])
ax2.set_xlim(zoom_start1, zoom_end1)
ax2.set_ylim(-1, 1)
ax2.tick_params(axis='x', colors=colors['black'])
ax2.tick_params(axis='y', colors='black')
ax2.grid(True, color=colors['white'])

# Define the second zoom region (38 to 38.5 µs)
zoom_start2 = 38
zoom_end2 = 38.5
zoom_indices2 = np.where((t >= zoom_start2) & (t <= zoom_end2))
t_zoom2 = t[zoom_indices2]
signal_zoom2 = signal[zoom_indices2]
ax1.axvspan(zoom_start2, zoom_end2, color=colors['light_green'], alpha=0.3)

# Plot 3: Amplitude vs Time (Zoomed-In for 38 to 38.5 µs)
ax3 = plt.subplot(3, 3, 5, facecolor=colors['soft_grey'])
ax3.plot(t_zoom2, signal_zoom2, color=colors['muted_blue'])
ax3.set_xlabel('Time (µs)', color=colors['black'])
ax3.set_ylabel('Amplitude', color=colors['black'])
ax3.set_title('Zoomed-In (38 to 38.5 µs)', color=colors['black'])
ax3.set_xlim(zoom_start2, zoom_end2)
ax3.set_ylim(-1, 1)
ax3.tick_params(axis='x', colors=colors['black'])
ax3.tick_params(axis='y', colors='black')
ax3.grid(True, color=colors['white'])

# Plot 4: Frequency vs Time
ax4 = plt.subplot(3, 3, 6, facecolor=colors['soft_grey'])
ax4.plot(t, instantaneous_frequency, color=colors['muted_blue'])
ax4.set_xlabel('Time (µs)', color=colors['black'])
ax4.set_ylabel('Frequency (GHz)', color=colors['black'])
ax4.set_title('Frequency vs Time', color=colors['black'])
ax4.set_xlim(t_start, t_end)
ax4.set_ylim(f_start, f_end)
ax4.tick_params(axis='x', colors=colors['black'])
ax4.tick_params(axis='y', colors='black')
ax4.grid(True, color=colors['white'])

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the color palette
colors = {
    'black': '#000000',
    'yellow': '#F3E079',
    'soft_grey': '#D9D9D9',
    'muted_blue': '#A7BBC7',
    'blue': 'blue',
    'white': '#FFFFFF',
    'beige': '#F5F2E0'
}

# Define the parameters
f_start = 77  # Start frequency in GHz
f_end = 81    # End frequency in GHz
T_chirp = 40  # Chirp duration in µs
B = f_end - f_start  # Bandwidth in GHz
S = B / T_chirp  # Chirp slope in GHz/µs
tau = 5  # Time delay in µs for RX chirp

# Time vector
t = np.linspace(0, T_chirp, 1000)  # Time in µs
t_rx = t + tau  # Time vector for RX chirp

# Frequency equations
tx_chirp = f_start + S * t  # TX chirp
rx_chirp = f_start + S * (t - tau)  # RX chirp

# Create the plot
plt.figure(figsize=(10, 8), facecolor=colors['beige'])

# Plot TX chirp
plt.plot(t, tx_chirp, label="TX chirp", color=colors['blue'], linewidth=1)

# Plot RX chirp
plt.plot(t_rx, tx_chirp, label="RX chirp", color=colors['blue'], linewidth=1, linestyle="--")

# Annotations for Slope S and Delay τ
plt.annotate("", xy=(tau, f_start + S*tau), xytext=(0, f_start + S*tau),
             arrowprops=dict(arrowstyle="<->", color=colors['black']))
plt.text(tau / 2, f_start + S*tau + 0.05, r"$\tau$", fontsize=12, ha="center", color=colors['black'])

plt.annotate("", xy=(tau, f_start), xytext=(tau, f_start + S * tau),
             arrowprops=dict(arrowstyle="<->", color=colors['black']))
plt.text(tau + 1, f_start + S * tau / 2, r"$S\tau$", fontsize=12, ha="center", rotation=90, color=colors['black'])

# Add labels and lines for T_c
plt.annotate("", xy=(0, f_end +0.05), xytext=(T_chirp, f_end+0.05 ),
             arrowprops=dict(arrowstyle="<->", color=colors['black']))
plt.text(T_chirp / 2, f_end + 0.1, r"$T_c$", fontsize=12, ha="center", color=colors['black'])

# Set plot background color
plt.gca().set_facecolor(colors['soft_grey'])

# Axis labels and ticks
plt.xlabel("Time ($\mu$s)", fontsize=14, color=colors['black'])
plt.ylabel("Frequency (GHz)", fontsize=14, color=colors['black'])
plt.xticks([0, tau, T_chirp], labels=["$0$", r"$\tau$", r"$T_c$"], color=colors['black'])
plt.yticks([f_start, f_start + S * tau, f_end],
           labels=["$f_c$", r"$f_c + S\tau$", "$f_c + B$"], color=colors['black'])

# Add title
plt.title("TX and RX Chirps with Delay", fontsize=14, color=colors['black'])

# Set limits
plt.xlim(0, T_chirp + 10)
plt.xlim(0, T_chirp + 5)  # Start X-pltis at 0
plt.ylim(f_start, f_end + 0.7)  # Start Y-axis at 77 GHz

# Add grid
plt.grid(color=colors['muted_blue'], linestyle="--", linewidth=0.5)

# Add legend
plt.legend(facecolor=colors['beige'], edgecolor=colors['soft_grey'])

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the color palette
colors = {
    'black': '#000000',
    'yellow': '#F3E079',
    'soft_grey': '#D9D9D9',
    'muted_blue': '#A7BBC7',
    'blue': 'blue',
    'white': '#FFFFFF',
    'beige': '#F5F2E0'
}

# Parameters
f_start = 77  # Start frequency in GHz
f_end = 81    # End frequency in GHz
T_chirp = 40  # Chirp duration in µs
B = f_end - f_start  # Bandwidth in GHz
S = B / T_chirp  # Chirp slope in GHz/µs

# Object delays (in µs)
delays = [5, 9, 15]  # Two objects with tau1 and tau2 delays

# Time vector
t = np.linspace(0, T_chirp, 1000)  # Time in µs

# TX Chirp
tx_chirp = f_start + S * t

# RX Chirps for multiple objects
rx_chirps = [f_start + S * (t - tau) for tau in delays]

# IF signals (constant tones due to delays)
if_signals = [S * tau * np.ones_like(t[t >= tau]) for tau in delays]
if_time = [t[t >= tau] for tau in delays]  # Time vectors for IF signals starting at tau

# Create the figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10), facecolor=colors['beige'])

# --- Top Plot: TX and RX chirps ---
axs[0].plot(t, tx_chirp, label="TX chirp", color=colors['blue'], linewidth=1)

for i, rx_chirp in enumerate(rx_chirps):
    axs[0].plot(t, rx_chirp, linestyle="--", label=f"RX Chirp ($\\tau_{i+1}$)", color=colors['muted_blue'])

# Vertical lines for delays and T_c
for tau in delays:
    axs[0].axvline(tau, color=colors['soft_grey'], linestyle="-", linewidth=0.5)
    axs[0].axvline(tau, color=colors['black'], linestyle="--", linewidth=0.5)
axs[0].axvline(T_chirp, color=colors['black'], linestyle="--", linewidth=0.5)  # T_c

# Horizontal line at f_c + B
axs[0].axhline(f_end, color=colors['black'], linestyle="--", linewidth=0.5)





# Top plot settings
axs[0].set_xlabel("Time ($\mu$s)", fontsize=12, color=colors['black'])
axs[0].set_ylabel("Frequency (GHz)", fontsize=12, color=colors['black'])
axs[0].set_xlim(0, T_chirp + 5)
axs[0].set_ylim(f_start, f_end + 2)
axs[0].set_yticks([f_start, f_end])  # Two y-ticks only
# axs[0].legend(loc="upper left", fontsize=10, facecolor=colors['beige'], edgecolor=colors['soft_grey'])
axs[0].grid(color=colors['soft_grey'], linestyle="-", linewidth=0.5)
axs[0].set_facecolor(colors['soft_grey'])
axs[0].set_xticks([0, 5, 9, 15, 40], labels=["$0$", r"$\tau_1=5$", r"$\tau_2=9$", r"$\tau_3=15$", r"$T_c=40$"], color=colors['black'])
axs[0].set_yticks([f_start, f_end], labels=[r"$f_c=77$", r"$f_c+B=81$"], color=colors['black'])

# --- Bottom Plot: IF signals ---
for i, (if_signal, if_t) in enumerate(zip(if_signals, if_time)):
    axs[1].plot(if_t, if_signal, label=f"IF Signal ($\\tau_{i+1}$)", color=colors['blue'], linewidth=1)

# Add delays as horizontal double-arrow lines
for i, tau in enumerate(delays):
    axs[1].annotate("", xy=(0, S * tau), xytext=(tau, S * tau),
                    arrowprops=dict(arrowstyle="<->", color=colors['black'], linewidth=1))
    axs[1].text(tau / 2, S * tau + 0.1, f"$\\tau_{i+1}$", fontsize=10, color=colors['black'], va="center")

# Add vertical grid lines at delays
for tau in delays:
    axs[1].axvline(tau, color=colors['black'], linestyle="--", linewidth=0.5)
    axs[1].axhline(S*tau, color=colors['black'], linestyle="--", linewidth=0.5)
axs[1].axvline(T_chirp, color=colors['black'], linestyle="--", linewidth=0.5)
# Bottom plot annotations and settings

axs[1].set_xlabel("Time ($\mu$s)", fontsize=12, color=colors['black'])
axs[1].set_ylabel("Frequency (GHz)", fontsize=12, color=colors['black'])
axs[1].set_xlim(0, T_chirp + 5)
axs[1].set_ylim(0, S * max(delays) + 1)
axs[1].set_yticks([S * delays[0], S * delays[1], S * delays[2]])  # Two y-ticks only
# axs[1].legend(loc="upper left", fontsize=10, facecolor=colors['beige'], edgecolor=colors['soft_grey'])
axs[1].set_facecolor(colors['soft_grey'])
axs[1].set_xticks([0, 5, 9, 15, 40], labels=["$0$", r"$\tau_1=5$", r"$\tau_2=9$", r"$\tau_3=15$", r"$T_c=40$"], color=colors['black'])
axs[1].set_yticks([0, S * delays[0], S * delays[1], S * delays[2]], labels=["$0$", r"$S\tau_1=0.5$", r"$S\tau_2=0.9$", r"$S\tau_3=1.5$"], color=colors['black'])
# Tight layout and show
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the color palette
colors = {
    'black': '#000000',
    'yellow': '#F3E079',
    'soft_grey': '#D9D9D9',
    'muted_blue': '#A7BBC7',
    'blue': 'blue',
    'green': 'green',
    'white': '#FFFFFF',
    'beige': '#F5F2E0'
}

# Parameters
f_start = 10  # Start frequency in Hz
f_end = 81  # End frequency in Hz
T_chirp = 20  # Chirp duration in seconds
B = f_end - f_start  # Bandwidth in Hz
S = B / T_chirp  # Chirp slope in Hz/s
tau_1 = 5  # Time delay in seconds for RX signal 1
tau_2 = 5.3  # Time delay in seconds for RX signal 2

# Sampling parameters
fs = 2 * f_end  # Sampling frequency in Hz (Nyquist rate)
t = np.linspace(0, T_chirp, int(fs * T_chirp))  # Time vector for TX chirp

# Generate TX chirp
tx_chirp = np.sin(2 * np.pi * (f_start * t + 0.5 * S * t**2))  # TX chirp

# Generate RX chirps for each delay
rx_chirp_1 = np.sin(2 * np.pi * (f_start * (t - tau_1) + 0.5 * S * (t - tau_1)**2))  # RX chirp for tau_1
rx_chirp_2 = np.sin(2 * np.pi * (f_start * (t - tau_2) + 0.5 * S * (t - tau_2)**2))  # RX chirp for tau_2

# IF signals (product of TX and RX chirps)
if_signal_1 = tx_chirp * rx_chirp_1
if_signal_2 = tx_chirp * rx_chirp_2

# Beat IF signals (cosine components due to the delays)
beat_if_signal_1 = 0.5 * (np.cos(2 * np.pi * S * tau_1 * t))
beat_if_signal_2 = 0.5 * (np.cos(2 * np.pi * S * tau_2 * t))

# Compute FFT of the combined IF signal
n = len(if_signal_1)  # Number of samples
fft_if_signal_1 = np.fft.fft(if_signal_1)
fft_if_signal_2 = np.fft.fft(if_signal_2)
frequencies = np.fft.fftfreq(n, d=1 / fs)  # Frequency axis in Hz

# Take the positive half of the spectrum
positive_freqs = frequencies[:n // 2]   # Convert to MHz
positive_spectrum_1 = np.abs(fft_if_signal_1[:n // 2])  # Magnitude spectrum for tau_1
positive_spectrum_2 = np.abs(fft_if_signal_2[:n // 2])  # Magnitude spectrum for tau_2



# Set the figure background
plt.figure(figsize=(12, 12), facecolor=colors['beige'])

# Plot IF signals in time domain
ax1 = plt.subplot(3, 1, 1, facecolor=colors['soft_grey'])
ax1.plot(t, beat_if_signal_1, color=colors['blue'], linewidth=1.5, label=f"IF Signal (τ₁={tau_1} s)", alpha = 0.5)
ax1.plot(t, beat_if_signal_2, color=colors['green'], linewidth=1.5, label=f"IF Signal (τ₂={tau_2} s)", alpha = 0.5)
ax1.set_title("IF Signal in Time Domain", fontsize=16, color=colors['black'])
ax1.set_xlabel("Time ($\mu$s)", fontsize=14, color=colors['black'])
ax1.set_ylabel("Amplitude", fontsize=14, color=colors['black'])
ax1.grid(color=colors['muted_blue'], linestyle="--", linewidth=0.5)
ax1.legend(fontsize=12, facecolor=colors['beige'], edgecolor=colors['soft_grey'])

# Plot IF signal frequency spectrum
ax2 = plt.subplot(3, 1, 2, facecolor=colors['soft_grey'])
ax2.plot(positive_freqs, positive_spectrum_1, color=colors['blue'], linewidth=1, label=f"Frequency Spectrum (τ₁={tau_1} s)")
ax2.plot(positive_freqs, positive_spectrum_2, color=colors['green'], linewidth=1, label=f"Frequency Spectrum (τ₂={tau_2} s)")
ax2.set_title("IF Signal Frequency Spectrum", fontsize=16, color=colors['black'])
ax2.set_xlabel("Frequency (MHz)", fontsize=14, color=colors['black'])
ax2.set_ylabel("Amplitude", fontsize=14, color=colors['black'])
ax2.grid(color=colors['muted_blue'], linestyle="--", linewidth=0.5)
ax2.set_xlim(0, 2*S*tau_2)
ax2.legend(fontsize=12, facecolor=colors['beige'], edgecolor=colors['soft_grey'])

plt.tight_layout()
plt.show()
```
